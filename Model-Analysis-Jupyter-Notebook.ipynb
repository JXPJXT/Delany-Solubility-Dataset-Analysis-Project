# **My First ML Project**

## 1.Load Data

```python
import pandas as pd
import sklearn
df = pd.read_csv("C:/Users/bhati/OneDrive/Desktop/ProjectML/dataset.csv")
print(df)
print(df.columns)
```

## 2.Data Preparation

### 2.1.Data Seperation as X and Y.

```python
#print(df.columns)
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1)
x
```

### 2.2.Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=100)
```

```python
X_train
```

```python
X_test
```

## 3.Model Building

### 3.1.Linear Regression

#### 3.1.1.Training The Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

#### 3.1.2.Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_test_pred
```

#### 3.2.3.Evalutate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
lr_results = pd.DataFrame({
    'Method': ['LinearRegression'],
    'Training MSE': [lr_train_mse],
    'Training R2': [lr_train_r2],
    'Test MSE': [lr_test_mse],
    'Test R2': [lr_test_r2]
})
```

```python
lr_results
```

### 3.2 Random Forest

#### 3.2.1.Training the Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train,y_train)
```

#### 3.2.2.Applying the Model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

#### 3.2.3.Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame({
    'Method': ['Random Forest'],
    'Training MSE': [rf_train_mse],
    'Training R2': [rf_train_r2],
    'Test MSE': [rf_test_mse],
    'Test R2': [rf_test_r2]
})
```

```python
rf_results
```

## 4.Model Comparsion

```python
df_models = pd.concat([lr_results,rf_results], axis=0).reset_index(drop=True)
df_models
```
